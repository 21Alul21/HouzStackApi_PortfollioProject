```python
# removing strings

df['Bedrooms'] = df.Bedrooms.apply(lambda x: x.split()[0])
df['Bathrooms'] = df.Bathrooms.apply(lambda x: x.split()[0])
df['Toilets'] = df.Bedrooms.apply(lambda x: x.split()[0])
df['Price'] = df.Price.apply(lambda x: x.split('/')[0])
```

```python
# removing row with value 'beds'
df.query('Bedrooms != "beds"', inplace=True)
```

```python
df.Bathrooms.unique()
```

```python
# removing row with value 'baths'
df.query('Bathrooms != "baths"', inplace=True)
```

```python
df.info()
```

```python
# converting object datatype to int for Bedroooms, Bathroooms, Toilets and Price

lst = ['Bedrooms', 'Bathrooms', 'Toilets']
for i in lst:
    df[i] = df[i].apply(lambda x: int(x))


# removing comma from Price Field
def comma_remover(x):
    number = ''
    for i in x:
        if i == ',':
            continue
        number += i
    return int(number)

df['Price'] = df['Price'].apply(comma_remover)
```

```python
# removing rows with prices greater than 1 billion
df.query('Price < 1000000000', inplace=True)
```

```python
X = df.drop(['Price'], axis=1)
```

```python
Y = df['Price']
```

```python
# splitting the data for training and testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

## Model Training

```python
xgboost = xgb.XGBRegressor()
```

```python
xgboost.fit(X_train, Y_train)
```

```python
prediction = xgboost.predict(X_test)
```
